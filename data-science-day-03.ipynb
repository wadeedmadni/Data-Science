# 1. Line Graph

```python
# import libraries
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# load IRIS dataset

df = sns.load_dataset("iris")
df
```

```python
# adding styles
# always add style before defining the x and y axis or else it won't work
sns.set_style(style=None, rc = None)
sns.set_style("dark")


sns.lineplot(x= "sepal_length", y = "sepal_width", data = df)

# adding title for the graph
plt.title("iris plot graph")

# adding x and y axis limits
# plt.xlim(2)
#plt.ylim(1)

# change figure size
plt.figure(figsize =(8,10))

plt.show()
```

# 1. Bar Plot  

```python
sns.barplot(x = "species", y = "petal_length", data = df)
plt.show()
```

**Grouping**

```python
titanic = sns.load_dataset("titanic")
titanic
```

```python
import numpy
sns.barplot(x = "class", y = "fare", hue = "sex", data = titanic, color = "orange", errorbar = None, palette = 'pastel', estimator = 'mean', saturation = 1)
plt.show()
```

```python
sns.barplot(x = "class", y = "fare", hue = "sex", data = titanic,
            linewidth = 3, errcolor = "0.5",
            color = "orange", errorbar = None, palette = 'pastel', 
            estimator = 'mean', saturation = 1)
```

# **2. Bar Plot**

```python
import seaborn
seaborn.set(style = 'whitegrid')

seaborn.boxplot(x = "class", y = "fare", data = titanic)
```

```python
import seaborn
seaborn.set(style = 'whitegrid')

tip = seaborn.load_dataset('tips')
tip.head(5)
```
