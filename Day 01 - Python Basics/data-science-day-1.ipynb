# **8. if, else Statements**

```python
school_eligibility = 4
high_school_eligibility = 10
child_age = input("What is your sibling's age?")
child_age = int(child_age)

# is your sibling eligible for school? (school_eligibility is 4)

if child_age >= school_eligibility and child_age < high_school_eligibility:
    print("Your sibling can go to school")
elif child_age >= high_school_eligibility:
    print("Your sibling should join High School")
else:
    print("Your sibling cannot go to school")
```
