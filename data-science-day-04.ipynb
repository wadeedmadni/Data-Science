# 2-D array

```python
a = np.array([
    [2,6,12,75,13],
    [5,75,45,932,1]
])
a
```

```python
b = np.array([
    [32,421,64,86,14],
    [24,64,23,81,131]
])
b
```

```python
c = np.concatenate((a,b), axis = 0)
c
```

```python
c = np.concatenate((a,b), axis = 1)
c
```
